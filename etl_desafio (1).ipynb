# Import data from .csv file (EXTRACT)

```python
import pandas as pd

df = pd.read_csv("music_sentiment.csv")
df.head()
```

# TRANSFORM

```python
df['fav_genre'] = df['fav_genre'].str.strip().str.title()
df["mood"] = df['mood'].str.strip().str.title()


df['prompt'] = df.apply(
    lambda row: f"Recommend 3 songs and a (max 10 words) description of each for someone who likes {row['fav_genre']} and is feeling {row['mood']}.",
    axis=1
)
```

```python

from openai import OpenAI
import os

client = OpenAI(
    api_key="insert api key",
    base_url="https://api.groq.com/openai/v1",
)

def gen_recommendations(prompt):
    response = client.responses.create(
        input=prompt,
        model="openai/gpt-oss-20b",
    )
    return response.output_text
```

#LOAD

Loading users and recomentations into a new csv

```python
df['recommendations'] = df['prompt'].apply(gen_recommendations)
df_final = df[['user_id', 'user', 'fav_genre', 'mood', 'recommendations']]
df_final.to_csv("music_recommendations.csv", index=False, encoding="utf-8")
```
